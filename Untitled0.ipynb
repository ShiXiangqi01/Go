```python
from sentence_transformers import SentenceTransformer

# 加载模型（首次运行会自动下载）
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# 生成向量
texts = ["感冒发烧"]
embedding = model.encode(texts)  # 输出numpy数组
print(embedding.shape)
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient(path="/db/chroma_demo") #similar database
collection = client.get_or_create_collection(name="my_collection") #similar table
#client.delete_collection(name="collection2")
collection2 = client.get_or_create_collection(name="collection2")

# documents = ["风寒感冒", "寒邪客胃", "心脾两虚"]
# ids = [str(uuid.uuid4()) for _ in documents]
ids = [str(uuid.uuid4()) for _ in chunk_list]
embeddings = [model.encode(chunk) for chunk in chunk_list]



collection2.add(
    ids = ids,
    documents = chunk_list,
    embeddings=embeddings
)

print(collection2.count())


```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# 模型名称（替换为实际蒸馏模型名称，如 "deepseek-ai/deepseek-moe-16b"）
model_name = "deepseek-ai/deepseek-moe-16b-base"

# 加载分词器和模型
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
# model = AutoModelForCausalLM.from_pretrained(
#     model_name,
#     torch_dtype=torch.float16,  # 半精度减少显存占用
#     device_map="auto",          # 自动分配 GPU/CPU
#     trust_remote_code=True
# )

# model.eval()  # 切换到推理模式
```

```python
 generation_model = AutoModelForCausalLM.from_pretrained( # Renamed to avoid conflict
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)
generation_model.eval()
```

```python
qs = "风热感冒"
qs_embedding = model.encode(qs)
res = collection2.query(
    query_embeddings = [qs_embedding],
    query_texts=qs,
    n_results = 2,

)

result = res['documents'][0]
context = "\n".join(result)
print(context)
```

```python
# 输入文本
prompt = f"""你是一个中医问答机器人，任务是根据参考信息回答用户问题，如果参考信息不足以回答用户问题，请回复不知道，不要去杜撰任何信息，请用中文回答。
    参考信息：{context}，来回答问题：{qs}，
    """

# 编码输入
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

# 生成文本
outputs = generation_model.generate(
    **inputs,
    max_new_tokens=100,  # 生成的最大长度
    temperature=0.7,     # 控制随机性（0~1，越小越确定）
    do_sample=True,
    use_cache = False, # 启用随机采样
)

# 解码输出
answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(answer)
```
